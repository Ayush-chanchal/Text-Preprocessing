```python
import pandas as pd
data=pd.read_csv("./IMDB Dataset.csv")
```

```python
dataset=data.iloc[:20,:]
print(dataset)
```

```python
dataset['review'].str.lower() # 1st step of preprocessing is lowercasing 
```

```python
#removing the punctuation from the dataset
import string
exclude= string.punctuation

def remove_punc1(text):
  return text.translate(str.maketrans('','',exclude))

dataset['review'].apply(remove_punc1)
```

```python
# Removing stopwords from the reviews of the dataset
from nltk.corpus import stopwords
stopwords.words('english')
```

```python
# Removing stopwords
def remove_stopwords(text):
  new_text=[]
  for word in text.split():
    if word in stopwords.words('english'):
      new_text.append('')
    else:
      new_text.append(word)
  x=new_text[:]
  new_text.clear()
  return " ".join(x)
```

```python
dataset['review'].apply(remove_stopwords)
```

```python
# Tokenisation
import spacy

nlp = spacy.load("en_core_web_sm")
def tokenize_review(review):
    doc = nlp(review)
    tokens = [token.text for token in doc]
    return tokens
# Tokenize each review in the dataset
tokenized_reviews = []
for review in dataset['review']:
    tokens = tokenize_review(review)
    tokenized_reviews.append(tokens)
dataset['Tokenized_Review'] = tokenized_reviews
print(dataset)
```

```python
# stemming (Stemming is faster than lemmatization.)
from nltk.stem.porter import PorterStemmer

ps=PorterStemmer()
def stem_words(text):
    return " ".join([ps.stem(word) for word in text.split()])
```

```python
dataset['review'].apply(stem_words)
```

```python
# Lemmatization
import nltk
from nltk.stem import WordNetLemmatizer
wordnet_lemmatizer = WordNetLemmatizer()

dat=nltk.word_tokenize(dataset['review'])
punctuation="?:;!,."
for word in dat:
    if word in punctuation:
        dat.remove(word)
dat
print("{0:20}{1:20}".format("word","lemma"))
for word in dat:
    print("{0:20}{1:20}".format(word,wordnet_lemmatizer.lemmatize(word,pos='v')))
```
